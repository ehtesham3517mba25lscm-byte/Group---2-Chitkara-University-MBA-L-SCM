```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the data
df = pd.read_csv('Cleaned_Maritime_Data.csv')
```

```python
# 2. Summary Statistics (The "General View")
# This provides the mean, median, and variance of your core metrics
summary_stats = df[['Median_Port_Time', 'Avg_Age', 'Avg_Size_GT', 'Avg_Cargo_DWT']].describe()
print("Descriptive Statistics:")
print(summary_stats)
```

```python
# 3. Correlation Analysis (The "Strength of Relationships")
# This shows if Median_Port_Time moves in sync with other variables
# Values closer to 1 or -1 indicate a stronger relationship
correlation_matrix = df[['Median_Port_Time', 'Avg_Age', 'Avg_Size_GT', 'Avg_Cargo_DWT', 'Avg_TEU']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Statistical Correlation: Factors Impacting Port Time')
plt.show()
```

```python
# 4. Statistical Significance Test (P-Value Check)
# We test if the mean Port Time for 'Container' vs 'Tanker' is statistically different
# A p-value < 0.05 indicates the difference is NOT due to chance
group1 = df[df['Ship_Type'] == 'Container']['Median_Port_Time']
group2 = df[df['Ship_Type'] == 'Tanker']['Median_Port_Time']
t_stat, p_value = stats.ttest_ind(group1, group2, nan_policy='omit')
print(f"\nStatistical Significance Test (T-test):")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
```
